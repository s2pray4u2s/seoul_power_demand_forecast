```python
import pandas as pd
import glob

file_list = sorted(glob.glob("seoul_data*.csv"))

merged_df = pd.read_csv(file_list[0], encoding="euc-kr", skiprows=9)
merged_df.columns = merged_df.columns.str.strip()
merged_df.columns = merged_df.columns.str.replace('\ufeff','')

for file in file_list[1:]:
    df = pd.read_csv(file, encoding="euc-kr", skiprows=9, sep=",")
    df.columns = df.columns.str.strip()
    df.columns = df.columns.str.replace('\ufeff','')
    print(df.columns.tolist())
    break

    merged_df = pd.merge(merged_df, df, on="일시", how="outer")

merged_df = merged_df.sort_values("일시")

merged_df.to_csv("seoul_horizontal_merged.csv", index=False, encoding="utf-8-sig")
print("병합 완료")
print(merged_df.head())
```
